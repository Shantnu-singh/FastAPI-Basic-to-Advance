```python
import numpy as np

def calculate_normalized_rmse(df, actual_col, predicted_col):
    """
    Calculate the normalized RMSE for the given dataframe and columns.

    Parameters:
    df (pd.DataFrame): The dataframe containing the data.
    actual_col (str): The column name for actual values.
    predicted_col (str): The column name for predicted values.

    Returns:
    float: The normalized RMSE value.
    """
    actual = df[actual_col]
    predicted = df[predicted_col]
    
    # Calculate RMSE
    rmse = np.sqrt(np.mean((actual - predicted) ** 2))
    
    # Normalize RMSE
    normalized_rmse = rmse / (actual.max() - actual.min())
    
    return normalized_rmse

# Example usage
normalized_rmse = calculate_normalized_rmse(df, 'ActualBaggageWeight', 'PredictedBaggageWeight')
print("Normalized RMSE:", normalized_rmse)
```
